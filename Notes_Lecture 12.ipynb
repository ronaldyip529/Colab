2. Line Graph


```python
# To plot a line graph
pit.plot( x_coords, y_coords)

# To include a title of the graph
pit.title("Title Caption")

# To include the x-axis label and y-axis label
pit.xlabel("Year")
pit.ylabel("Sales Amount")

# To display the graph
pit.show()

# To set the axis limit
pit.xlim(xmin=1, xmax=10)
pit.ylim(xmin=1, xmax=6)

# To show the grid line
pit.grid(True)

# To customize the tick mark’s label
pit.xticks( [ 0, 1, 2], ["Base Ball", "Basketball", "Football"] )
pit.yticks( [ 0, 1, 2, 3], ["Zero", "Quarter", "Half", "Three Quarter"] )

# To display markers at the Data Points
pit.plot(x_coords, y_coords, marker="o")
```

```python
import matplotlib.pyplot as pit
def main():
  x_coords = [0,1,2,3,4]
  y_coords = [0,3,1,5,2]
# Build the line graph
  pit.plot(x_coords, y_coords, marker="o")
# Add a title
  pit.title('Sales by Year')
# Add lables to the axes
  pit.xlabel("Year")
  pit.ylabel("Sales")
# Set the tick marks
  pit.xticks( [0, 1, 2, 3, 4], ['2016', '2017', '2018', '2019', '2020'] )
  pit.yticks( [0, 1, 2, 3, 4, 5], ['$0m', '$1m', '$2m', '$3m', '$4m', '$5m'] )
# Add a grid
  pit.grid(True)
# Display the chart
  pit.show()

main()
```

```python
import matplotlib.pyplot as pit
def main():
# Create a list with the X coordinates of each bar's left edge
  left_edges = [0, 10, 20, 30, 40]
# Create a list with the heights of each bar
  heights = [100, 200, 300, 400, 500]
# Create a variable for the bar width
  bar_widths = 10
  # Build the bar chart
  pit.bar(left_edges, heights, bar_widths,
  color=('r', 'g', 'b', 'w', 'k'))
# Add a title
  pit.title('Sales by Year')
# Add lables to the axes
  pit.xticks( [5, 15, 25, 35, 45],
['2016', '2017', '2018', '2019', '2020'] )
  pit.yticks( [0, 100, 200, 300, 400, 500],
['$0m', '$1m', '$2m', '$3m', '$4m', '$5m'] )
# Display the chart
  pit.show()

main()
```

4. Pie Chart

```python
# To plot a pie chart
pit.pie(sales)
# To create a list of the label
pit.pie( sales, labels=slice_labels)
# To change the color of the slices
pit.pie( sales, colors=["color1", "color2", …] )
# To include the percentage
pit.pie(sales, labels=slice_lables, autopct='%.2f%%')
```

```python
import matplotlib.pyplot as pit
def main():
# Create a list of sales amounts
  sales = [100, 400, 300, 600]
# Create a list of labels for the slices
  slice_lables = ['1st Qtr', '2nd Qtr', '3rd Qtr', '4thQtr']
  # Create a pie chart from the values
# pit.pie(sales, labels=slice_lables)
  pit.pie(sales, labels=slice_lables,
  autopct='%.2f%%',
  colors={'r', 'g', 'y', 'b'})
# Display the chart
  pit.show()
main()
```
