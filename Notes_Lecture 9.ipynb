File handling
1. An important activity in every application.
2. Files reside in the computer's secondary memory,
3. Access mode for an opened file controls the types of activities that can be performed.
4. A file handle indicates where data should be read or put into the file.

Access Modes
1. Read only (‘r’)
2. Read and Write (‘r+’)
3. Write only (‘w’)
4. Write and Read (‘w+’)
5. Append only (‘a’)
6. Append and Read (‘a+’)
7. Create (‘x’)

OPENING FILES
1. open() method - signals to the operating system to search for the file by its name and ensure that it exists.

```python
# Asking user to enter a filename
fname = input('Enter the file name: ')
fhand = open(fname)
count = 0
for line in fhand:
  count = count + 1
  print('There are', count, 'lines in', fname)
```

WRITING TO A FILE


write() method – inserts a string into the opened text file on a single line.


```python
#write() method – inserts a string into the opened text file on a single line.
f.write("Hello There\n")
```

```python
#writelines() method – inserts multiple string elements in a list into the opened text file.
f.writelines(["Hello World ", "You are welcome! \n"])
```

```python
# This program shows how to write data in a text file.
file = open("myfile.txt","w")
L = ["This is Lagos \n","This is Python \n","This is Fcc \n"]
# i assigned ["This is Lagos \n","This is Python \n","This is Fcc \n"] to
# variable L, you can use any letter or word of your choice.
# Variables are containers in which values can be stored.
# The \n is placed to indicate the end of the line.
file.write("Hello There \n")
file.writelines(L)
file.close()
```

```python
# read(n) method – returns the bytes read as a string. If no n is specified, it then reads the entire file.
f = open("myfiles.txt", "r") # ’r’ opens the text file for reading only
print(f.read()) # prints out the data in the text file
```

```python
"""readline(n) method
– reads a line from a file and returns a string. It reads at most n bytes if n is given.
But even if n is greater than the length of the line, it does not read more than one line."""
f = open("myfiles.txt", "r")
print(f.readline())
```

```python
readlines() method – reads all the lines and returns them as string elements in a list, one for each line.
```

```python
# repr() method – to check for the special characters
# Get the file handler
fhand = open('daffodils.txt')
# Loop through each line via file handler
for line in fhan:
print(repr(line))
```

```python
print() method with the default end value of print alerted
# Get the file handler
fhand = open('daffodils.txt')
# Loop through each line and modify the default value of 'end'
for line in fhand:
print(line, end = '')
```

```python
# rstrip(), right-strip, method – removes certain characters around a string
# Get the file handler
fhand = open('daffodils.txt')
# Loop through each line and remove extra line character with rstrip()
for line in fhand:
line = line.rstrip()
print(line)
```

try-except block – handles the case when the file to be opened does not exist. It makes the program terminate gracefully.

```python
fname = input('Enter the file name: ')
try:
  fhand = open(fname)
except:
  print('File nout found and can not be opened:', fname)
  exit()
count=0
for line in fhand:
  count = count + 1
print('There are', count, 'lines in', fname)
```
