CREATING A DIRECTORY

```python
# os.mkdir method
# FileExistError will be thrown if the directory exists
import os
if not os.path.isdir("program_source"):
  os.mkdir("program_source")

os.chdir("program_source”)
if not os.path.isdir("sample"):
  os.mkdir("sample")
```

```python
#os.rmdir() method – to remove a directory if exists. It deletes a directory only if it is empty. If the directory has contents, Python returns the error OSError: Directory not empty.

import os
os.rmdir("demo")
```

```python
# shutil.retree() method – to remove a directory and all its contents. The deletion is immediate and permanent, so use it with great care.

import shutil
shutil.rmtree("demo")
```

# Rename, Move and Copy Files and Directories

1.  shutil.rename() method
- to change the name of a file or a directory or to move it to a different location by giving the appropriate directory.

2. shutil.remove() method
- to delete a file.

3. shutil.copy() method
- to copy a source file to the specified destination and preserve the file’s metadata in the copy.

4. shutil.copyfile() method
- to copy a source file to the specified destination but not preserve the file’s metadata in the copy.

```python
import shutil
shutil.copy("myfile.txt", "copy.txt")

shutil.rename("copy.txt", "spare.txt")

os.mkdir("today")
shutil.copy("myfile.txt", "today") # copy to the today directory

shutil.copytree("today", "backup") # copy today directory and its content
shutil.move("files", "archive/files") # move files directory to archive/files directory
# if not exists, creates the directory
```
