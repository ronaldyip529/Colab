3. Constructor Method
- A constructor method is used to create and initialize an object from the class template. In Python, "__init__" is used
- Use the keyword class
- Use a capitalized letter for the class name


```python
class Pet():
  def __init__(self, height):
    self.height = height #an attribute of the object

  is_human = False
  owner = "David Wong"


#Usage of the Pet class
pet1 = Pet (height=5)
print("Pet's owner:",pet1.owner)
print("Pet's human status:", pet1.is_human)
print("Pet's height: ", pet1.height)
```

4. Object Properties
- Properties are used to manage the attributes of objects

5. Custom Methods for the Class

Three types of custom methods:

a) Instance Method;

b) Static Method;

c) Class Method

```python
"""
a) Instance Methods
- can assess the instance of a class
- read and write instance attributes
"""

class Country():
  def __init__(self, name= "Unspecified", population=None, size_kmsq=None):
    self.name = name
    self.population = population
    self.size_kmsq = size_kmsq
  def size_miles_sq(self, conversion_rate = 0.621371):
    return self.size_kmsq * conversion_rate ** 2

usa =Country(name= "United States of America", size_kmsq=9.8e6)
usa.size_miles_sq()
```

```python
class City():
  def __init__(self, name= "Unspecified", population=None, size_kmsq=None):
    self.name = name
    self.population = population
    self.size_kmsq = size_kmsq
  def __str__(self):
    label = self.name
    if self.population:
      label = "%s, population: %s"%(label, self.population)
    if self.size_kmsq:
      label = "%s, size_kmsq: %s"%(label, self.size_kmsq)
    return label
# Create the object
hksar= City(name="HKSAR", population=7e6)
print(hksar)
print('*'*50)
```

```python
"""
b) Static Method
- defined by using the @staticmethod decorator
- altering the behaviors of functions and classes
"""
import datetime
class Diary():
  def __init__(self, birthday, Christmas):
    self.birthday = birthday
    self.christmas = Christmas

'''# Method 1
  def format_date(self, date):
    return date.strftime("%d-%b-%y : %A")
'''
  @staticmethod
  def format_date(date):
    return date.strftime("%d-%b-%y : %A")
  def show_birthday(self):
    return self.format_date(self.birthday)
  def show_christmas(self):
    return self.format_date(self.christmas)

  # Create object
my_diary = Diary(datetime.date(2023, 5, 14), datetime.date(2023, 12, 25))
print("Birthday: ", my_diary.show_birthday())
print("Christmas Day: ", my_diary.show_christmas())
```

```python
"""c)_Class Method

- not bound to any specific instance. It is bound to the class only.
- The object class, (cls), is being passed as the first argument.
- Use @classmethod decorator.
"""
class Country():
  def __init__(self, name= "Unspecified", population=None, size_kmsq=None):
    self.name = name
    self.population = population
    self.size_kmsq = size_kmsq

  @classmethod
  def create_with_msq(cls, name, population, size_msq):
    size_kmsq = size_msq / 0.621371 ** 2
    return cls(name, population, size_kmsq)

mexico = Country.create_with_msq('Mexico', 150e6, 760000)
print(mexico.size_kmsq)
```

```python
class Person():
  def __init__(self, first_name, last_name):
    self.first_name = first_name
    self.last_name = last_name

  @property
  def full_name(self):
    return '%s %s ' % (self.first_name, self.last_name)
  @full_name.setter
  def full_name(self, name):
    first, last = name.split(' ')
    self.first_name = first
    self.last_name = last

# Construct the object
customer = Person("Lily", "Wong")
customer.full_name = "Lily Wong" # valid
print(customer.full_name)
```
