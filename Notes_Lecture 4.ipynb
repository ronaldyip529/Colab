# **LIST**

1. Stored multi data sets at the same time
2. Similar to array in other programmes
3. Allow multiple types in a list
4. Indexed, with definite positions
5. Ordered when it is set
6. Mutable, the variables can be changed
7. Change its elements by replacing, adding and removing items
8. Can be executed in for loop
9. Square brackets
10. Can be nested, sllow multiple data structures, two-dimensional array
11. Allow duplicate values

```python
shopping = ["bread", "milk", "eggs"]
print(shopping)

for item in shopping:
  print(item)
```

NESTED LIST

1. A list item can have another list structure.
2. Useful to present a table structure (i.e. two-dimensional array)

```python
m = [[1,2,3], [4,5,6]]
print(m[1][1]) # 5
for i in range(len(m)):
  for j in range(len(m[1])):
      print(m[i][j])
```

LIST METHODS

▪ A set of list methods for storing and retrieving values in the list

**Common** **Operations**
1. Slicing Data
2. Sorting Data
2. Appending Data
3. Searching Data
4. Inserting Data
5. Removing Data

```python
# append()
list = ["apple", "orange", "kiwifruit"]
list.append("pineapple")
print(list)
```

```python
# clear()
list = ["apple", "orange", "kiwifruit"]
list.clear()
print(list)
```

```python
# copy()
list = ["apple", "orange", "kiwifruit"]
copylist = list.copy()
print(copylist)
```

```python
# count()
list = ["apple", "orange", "kiwifruit"]
print(list.count("apple"))
```

```python
# extend()
list = ["apple", "orange", "kiwifruit"]
list1 = ["rice", "noodles"]
list.extend(list1)
print(list)
```

```python
# index()
list = ["apple", "orange", "kiwifruit"]
print(list.index("orange"))
```

```python
# insert()
fruits = ['apple', 'banana', 'cherry']
fruits.insert(1, "orange")
print(fruits)
```

```python
# pop()
fruits = ['apple', 'banana', 'cherry']
fruits.pop(1)
print(fruits)
```

```python
# remove()
fruits = ['apple', 'banana', 'cherry']
fruits.remove("apple")
print(fruits)
```

```python
# reverse()
fruits = ['apple', 'banana', 'cherry']
fruits.reverse()
print(fruits)
```

```python
# sort()
cars = ['Ford', 'BMW', 'Volvo']
cars.sort()
print(cars)
```

LIST CONCATENATION

```python
# Using + operator
list1 = [1, 2, 3]
list2 = [4, 5, 6]
final_list = list1 + list2
print(final_list)
```

LIST DUPLICATION

```python
# Using * operator
list3 = ["hello"]
print(list3*3)
```

LIST ELEMENT ACCESSING

```python
shopping = ["bread", "milk", "eggs"]
print(shopping[1])
shopping[1] = "banana"
print(shopping[1])
print(shopping[-1])
print(shopping[0:2])
print(shopping[:3])
print(shopping[1:])
```

DICTIONARY

1. Store a collection of data in key/value pairs
2. Dictionary is an unordered collection.
3. An ordered collection with a specific order that Python maintains.
4. Dictionary is written with curly brackets { } which have keys and values.
5. Keys must be unique
- no duplicate keys are allowed.
- Keys must be immutable
- they can be a string, a number, or a tuple.
6. Item value can be accessed with the key: dog2['breed']

```python
employee = {
"name": "Jack Nelson",
"age": 32,
"department": "Sales"
}
print(employee["age"])
```

SET

1. A set is a built-in Python data structure used to store a collection of unique items, potentially of mixed data types, in a single variable.
2. A set does not allow multiple occurrences of the same element.
3. A set is
- Unordered – the items of a set do not have any defined order.
- Unindexed – cannot access the items with [i] as with lists.
- Immutable – cannot change the existing items in the set, but can add further items as needed.
- Iterable – can loop over the items of a set.
- All items in the set must be immutable data types such as integers, floats, tuples, and strings. A set can be established using the set() constructor.
4. A set can also be initialized using curly brackets directly.
5. The number of items in a set can be obtained using the len() function.


```python
thisset = {"apple", "banana", "cherry", "apple"}
print(thisset)
```

```python
thisset = set("apple", "banana", "cherry", "apple")
print(thisset)
```
